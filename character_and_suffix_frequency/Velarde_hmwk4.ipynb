```python

import matplotlib.pyplot as plt

def ProcessFile():
    '''
    $ProcessFile()$ reads the file, and letter count for the types in the dictionary that was created to store the data
    
    PARAMETERS:
    @param none
    
    RETURN:
    @return {dict} (letter_counts) - creates a dictionary of letters and letter frequencies 
    
    @return {dictionary} (letter_counts) -the dictionary letter_counts is returned, by calculating the letter count for each letter type
    if the letter of line is in dictionary, increases the letter count for that type.
    '''
    # open the text file and read first line
    myFile = open("1200wsj-1.txt", "r")
    first_line = ''
    first_line = myFile.readline()
    first_line_lower = first_line.lower() 

    
    number_of_letters = len(first_line)
    
    #counts letters and their frequencies
    letter_counts = {}
    for char in list(first_line): # convert to list to get each character of line separated.
        if char in letter_counts:
            letter_counts[char] += 1
        else:
            letter_counts[char] = 1

            
    myFile.close()
    
    return letter_counts

def plotCharacterFrequency(letter_counts):
    '''
    $plotCharacterFrequency$ takes in dictionary letter_counts, and plots a character frequency vs word type graph. 
    The function creates a list of ordered characters, and corresponding ordered character frequencies from list of characters and frequencies.
    Finally, it plots the ordered data from the dictionary into a bar graph.
    
    
    PARAMETERS:
    @param {dict} letter_counts = This is a dictionary of letters, and frequency of that word in the txt file.
    
    RETURN:
    @return The function doesnt have a return, its plots a graph, and displays it.
    '''
    letters_and_frequencies = []
    for l in letter_counts:
        letters_and_frequencies.append((letter_counts[l], l))
    
    letters_and_frequencies.sort(reverse=True)
    
    #extracts frequencies and letters from letter_and_frequencies
    #creates usable data to plot
    ordered_frequency = []
    for i in letters_and_frequencies:
        freq = i[0]
        ordered_frequency.append(freq)
    
    ordered_letters = []
    for j in letters_and_frequencies:
        letter = j[1]
        ordered_letters.append(letter)
    
    #plots elements of ordered lists in descending order
    plt.figure(figsize=(10, 6)) # make bigger to fit data
    plt.bar(ordered_letters, ordered_frequency)
    plt.xlabel("characters")
    plt.ylabel("Frequency")
    
    plt.show()
    
def main():
    '''
    $main$ receives the output of ProcessFile() function, and inputs it into the function 'plotCharacterFrequency'
    that plots the data returned.
    
    PARAMETERS:
    @param- no params
    
    RETURN:
    @return The function doesnt have a return, it calls functions
    '''
    letter_counts = ProcessFile()
    plotCharacterFrequency(letter_counts)
main()
```

```python

import matplotlib.pyplot as plt

def ProcessFile():
    '''
    $ProcessFile()$ reads the file, and word count for the types in the dictionary that was created to store the data
    
    PARAMETERS:
    @param none
    
    RETURN:
    @return {dictionary} (suffix_counts) -the dictionary suffix_counts is returned, by calculating the suffix count for each word type
    if the suffix is in dictionary, increases the suffix count for that type.
    '''
    
    suffix_counts = {}
    # open the text file and read first line
    myFile = open("1200wsj-1.txt", "r")
    first_line = ''
    first_line = myFile.readline()
    first_line_lower = first_line.lower() 
    
    token_list = first_line.split()
    # list of words
    #2 length stems
   
    #iterate through each word in token list, extracting stems, and suffix, then comparing them to
    for i in range(len(token_list)):
        word = token_list[i]
        size = len(word)
        if len(word) > 3:
            stem2_substring = word[0:(size-2)] #stem
            suffix2_substring = word[(size-2):size] # ending (2)
    
            stem3_substring = word[0:(size-3)] #stem
            suffix3_substring = word[(size-3):size] #ending (3)
        
            
            if stem2_substring in first_line:
                #add suffix2_substring to dictionary with incrementing count if already in.
                if suffix2_substring in suffix_counts:
                    suffix_counts[suffix2_substring] += 1
                else:
                    suffix_counts[suffix2_substring] = 1
                
            if stem3_substring in first_line:
                #add suffix3_substring to dictionary
                if suffix3_substring in suffix_counts:
                    suffix_counts[suffix3_substring] += 1
                else:
                    suffix_counts[suffix3_substring] = 1
    
    myFile.close()
    return suffix_counts

def plotSuffixFrequency(suffix_counts):
    '''
    $plotSuffixFrequency$ takes in dictionary suffix_counts, and plots a suffix frequency vs suffix graph. 
    The function creates a list of ordered suffixes, and corresponding ordered suffix frequencies from list of suffixes and frequencies.
    Finally, it plots the ordered data from the dictionary into a bar graph.
    
    
    PARAMETERS:
    @param {dict} suffix_counts = This is a dictionary of words, and frequency of that word in the txt file.
    
    RETURN:
    @return The function doesnt have a return, its plots a graph, and displays it.

    '''
    words_and_frequencies = []
    for word in suffix_counts:
        words_and_frequencies.append((suffix_counts[word], word))
    
    words_and_frequencies.sort(reverse=True)
    
    #extracts frequencies and words from words_and_frequencies
    #creates usable data to plot
    ordered_frequency = []
    for i in words_and_frequencies:
        freq = i[0]
        ordered_frequency.append(freq)
    
    ordered_words = []
    for j in words_and_frequencies:
        word = j[1]
        ordered_words.append(word)
    
    #plots highest 20 elements of ordered lists
    plt.bar(ordered_words[:20], ordered_frequency[:20])
    plt.xlabel("suffix")
    plt.ylabel("Frequency")
    
    plt.show()
    
def main():
    '''
    $main$ receives the output of ProcessFile() function, and inputs it into the function 'plotCharacterFrequency'
    that plots the data returned.
    
    PARAMETERS:
    @param- no params
    
    RETURN:
    @return The function doesnt have a return, it calls functions
    '''
    suffix_counts = ProcessFile()
    plotSuffixFrequency(suffix_counts)
main()
```
